```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Load Dataset
df = pd.read_excel(r"C:\Users\MOHAMED-OSAMA\Desktop\Football_Dataset.xlsx")
```

```python
# Basic Preprocessing
df['MatchDate'] = pd.to_datetime(df['MatchDate'], errors='coerce')
```

```python
# Group by Team
# ================================
grouped = df.groupby('Team')
```

```python
# Create DataFrame
# ================================
team_stats = pd.DataFrame()
team_stats['Matches'] = grouped['MatchDate'].nunique()
```

```python
# Total goals scored
# ================================
team_stats['Total Goals'] = grouped['Goals'].sum()
```

```python
# Average goals per match
team_stats['Avg Goals per Match'] = team_stats['Total Goals'] / team_stats['Matches']
```

```python
# Goal conversion rate
team_stats['Conversion Rate'] = grouped['Goals'].sum() / grouped['Shots'].sum()
```

```python
# xG efficiency
team_stats['xG Efficiency'] = grouped['Goals'].sum() / grouped['xG'].sum()
```

```python
# Total assists
team_stats['Total Assists'] = grouped['Assists'].sum()
```

```python
# Chance creation
team_stats['Chance Creation'] = grouped['KeyPasses'].sum()
```

```python
# Assist efficiency
team_stats['Assist Efficiency'] = grouped['Assists'].sum() / grouped['KeyPasses'].sum()
```

```python
# Dribbles per match
team_stats['Dribbles per Match'] = grouped['Dribbles'].sum() / team_stats['Matches']
```

```python
# Average rating
team_stats['Avg Rating'] = grouped['Rating'].mean()
```

```python
# ================================
# Passing KPIs
# ================================
```

```python
# Average pass accuracy
team_stats['Avg Pass Accuracy'] = grouped['PassAccuracy'].mean()
```

```python
# Passing impact
team_stats['Passing Impact'] = grouped['KeyPasses'].sum() * team_stats['Avg Pass Accuracy']
```

```python
# ================================
# Defensive KPIs
# ================================

```

```python
# Defensive actions
team_stats['Defensive Actions'] = (
    grouped['Tackles'].sum() +
    grouped['Interceptions'].sum() +
    grouped['Clearances'].sum()
)
```

```python
# Defensive intensity
team_stats['Defensive Intensity'] = team_stats['Defensive Actions'] / team_stats['Matches']
```

```python
# ================================
# Discipline KPIs
# ================================
```

```python
# Fouls per match
team_stats['Fouls per Match'] = grouped['FoulsCommitted'].sum() / team_stats['Matches']
```

```python
# Yellow cards rate
team_stats['Yellow Cards Rate'] = grouped['YellowCards'].sum() / team_stats['Matches']
```

```python
# Red cards rate
team_stats['Red Cards Rate'] = grouped['RedCards'].sum() / team_stats['Matches']
```

```python
# Discipline score
team_stats['Discipline Score'] = -(
    grouped['YellowCards'].sum() +
    2 * grouped['RedCards'].sum() +
    grouped['FoulsCommitted'].sum()
)
```

```python
# ================================
# Efficiency KPIs
# ================================
```

```python
# Goal contribution
team_stats['Goal Contribution'] = (
    grouped['Goals'].sum() + grouped['Assists'].sum()
)
```

```python

# Contribution per match
team_stats['Contribution per Match'] = team_stats['Goal Contribution'] / team_stats['Matches']
```

```python
# Shot efficiency
team_stats['Shot Efficiency'] = grouped['Shots'].sum() / grouped['Goals'].sum()
```

```python
# xA efficiency
team_stats['xA Efficiency'] = grouped['Assists'].sum() / grouped['xA'].sum()
```

```python
# ================================
# Advanced KPIs
# ================================
```

```python
# Attacking index
team_stats['Attacking Index'] = (
    grouped['Goals'].sum()*4 +
    grouped['Assists'].sum()*3 +
    grouped['KeyPasses'].sum()*2 +
    grouped['Dribbles'].sum()
)
```

```python
# Defensive index
team_stats['Defensive Index'] = (
    grouped['Tackles'].sum()*2 +
    grouped['Interceptions'].sum()*2 +
    grouped['Clearances'].sum()
)
```

```python
# Team strength
team_stats['Team Strength'] = (
    team_stats['Attacking Index'] +
    team_stats['Defensive Index'] +
    team_stats['Avg Rating']
)
```

```python
# Risk score
team_stats['Risk Score'] = (
    grouped['FoulsCommitted'].sum() +
    grouped['YellowCards'].sum()*2 +
    grouped['RedCards'].sum()*3
)
```

```python
# ================================
# Deep Analytics KPIs
# ================================
```

```python
# Finishing quality
team_stats['Finishing Quality'] = grouped['Goals'].sum() / grouped['xG'].sum()
```

```python
# Shot quality
team_stats['Shot Quality'] = grouped['xG'].sum() / grouped['Shots'].sum()
```

```python
# Creative efficiency
team_stats['Creative Efficiency'] = grouped['Assists'].sum() / grouped['xA'].sum()
```

```python
# Chance utilization
team_stats['Chance Utilization'] = grouped['Goals'].sum() / grouped['KeyPasses'].sum()
```

```python
# Direct play
team_stats['Direct Play'] = grouped['Shots'].sum() / team_stats['Avg Pass Accuracy']
```

```python
# Build-up play
team_stats['Build Up Play'] = (
    (grouped['KeyPasses'].sum() + grouped['Assists'].sum()) *
    team_stats['Avg Pass Accuracy']
)
```

```python
# Dribble dependency
team_stats['Dribble Dependency'] = (
    grouped['Dribbles'].sum() /
    (grouped['Goals'].sum() + grouped['Assists'].sum())
)
```

```python
# Attack diversity
team_stats['Attack Diversity'] = (
    (grouped['Goals'].sum() +
     grouped['Assists'].sum() +
     grouped['KeyPasses'].sum() +
     grouped['Dribbles'].sum()) /
    grouped['Goals'].sum()
)
```

```python
# Defensive efficiency
team_stats['Defensive Efficiency'] = (
    (grouped['Tackles'].sum() + grouped['Interceptions'].sum()) /
    grouped['FoulsCommitted'].sum()
)
```

```python
# Pressure intensity
team_stats['Pressure'] = (
    (grouped['Tackles'].sum() + grouped['FoulsCommitted'].sum()) /
    team_stats['Matches']
)
```

```python
# Consistency
team_stats['Consistency'] = 1 / grouped['Rating'].std()
```

```python
# Goal stability
team_stats['Goal Stability'] = 1 / grouped['Goals'].std()
```

```python
# Offensive impact
team_stats['Offensive Impact'] = (
    (grouped['Goals'].sum()*4 +
     grouped['Assists'].sum()*3 +
     grouped['KeyPasses'].sum()*2 +
     grouped['Dribbles'].sum()) /
    team_stats['Matches']
)
```

```python
# Defensive impact
team_stats['Defensive Impact'] = (
    (grouped['Tackles'].sum()*2 +
     grouped['Interceptions'].sum()*2 +
     grouped['Clearances'].sum()) /
    team_stats['Matches']
)
```

```python
# Total impact
team_stats['Total Impact'] = (
    team_stats['Offensive Impact'] +
    team_stats['Defensive Impact'] -
    team_stats['Risk Score']
)
```

```python
# Overperformance
team_stats['Overperformance'] = (
    grouped['Goals'].sum() - grouped['xG'].sum()
)
```

```python
# Overperformance ratio
team_stats['Overperformance Ratio'] = (
    team_stats['Overperformance'] / grouped['xG'].sum()
)
```

```python
# Contribution gap
team_stats['Contribution Gap'] = (
    (grouped['Goals'].sum() + grouped['Assists'].sum()) -
    (grouped['xG'].sum() + grouped['xA'].sum())
)
```

```python
# Smart play
team_stats['Smart Play'] = (
    (grouped['KeyPasses'].sum() + grouped['Interceptions'].sum()) /
    grouped['FoulsCommitted'].sum()
)
```

```python
# Elite score
team_stats['Elite Score'] = (
    (grouped['Goals'].sum()*4 +
     grouped['Assists'].sum()*3 +
     grouped['KeyPasses'].sum()*2 +
     grouped['Dribbles'].sum()) +
    (grouped['Tackles'].sum() +
     grouped['Interceptions'].sum() +
     grouped['Clearances'].sum()) +
    team_stats['Avg Rating'] -
    (grouped['FoulsCommitted'].sum() +
     grouped['YellowCards'].sum()*2 +
     grouped['RedCards'].sum()*3)
)
```

```python
# ================================
# Final Output
# ================================
```

```python
team_stats = team_stats.replace([np.inf, -np.inf], np.nan)
team_stats = team_stats.fillna(0)

print(team_stats.head())
```
